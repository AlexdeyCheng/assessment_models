# AR

对时间上的随机变量建立线性相关方程。利用Yule-Walker方程，按确定步长k求解每一记忆参量的协方差，与t = 0处值作比得到$\phi_i$。高阶时引入Toeplizt矩阵方程求解$\phi_i$。$\epsilon_t$为残差值。
刻画方程为：

$y_t =\beta_0 + \sum_{i=1}^p \phi_i y_{t-i} + \epsilon_t $
