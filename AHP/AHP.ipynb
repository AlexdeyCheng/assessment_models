# AHP
    Analytic Hierarchy Process，即层次分析法。
## 方法核心
    将问题拆分为目标层，准则层和方案层。利用矩阵处理，对准则层的权重分配进行更科学、可量化、可解释的操作。同时，可以对方案层的打分给出矫正和一致性检验。可以将准则设置为子目标，从而实现嵌套的层次分析。

## 数学图景
    对于准则和方案评分，都给出preference方阵。比如准则之间的矩阵，a~ij~表示准则i比准则j重要a~ij~倍。比如方案在准则k上的评分矩阵，a~ij~表示在准则k上，方案i比方案j重要a~ij~倍。可以推导得出，当这些矩阵各行各列成比例时，评分是科学有效的。通过对比RI,CI等参数，可以判断矩阵是否近似有效（一致性检验，consistency test）。

    得到一致性检验通过的矩阵后，可以通过一定算法得到权重分配。常见的算法有几何平均，算术平均和特征值法。
    
    准则得到一个权重向量c^n，n为准则数量。各方案对各准则则得到al~i^k，k为方案数量，i = 1，2 , ......n .矩阵S = [al~i^k]。求W^k = S * c^n，得到一个包含各方案评分的向量。择优即可。

    有数学方法能将常见的打分表转化为preference矩阵，由此可以判断打分是否具有一致性和科学性。在此不赘述。

## 方法点评
    AHP能提供科学的权重计算，但是仍然依靠专家的打分和准则制定是否科学（只能评判，难以科学改进）
